```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#Generar datos para la distribución normal

media= 70 #Media

desv_std= 4 #Desviación estándar

num_datos= 500
```

```python
#generamos los datos de manera aleatoria y Lon ponemos en una Lista (array)

datos =np.random.normal(media, desv_std, num_datos)

datos[0:50] # Mostrar los primeros 50
```

```python
#Crear un rango de valores para crear una densidad

x= np.linspace(media-3*desv_std, media+3*desv_std, num_datos)

#Calcular La PDF de La distribución normal

pdf = 1/(desv_std*np.sqrt(2*np.pi))*np.exp(-(x-media)**2/(2*desv_std **2))
```

```python
#Para no imprimir toda la Lista de valores (500), desplegamos Los primeros 50

print("Imprimir los primeros 50 Valores de x", x[0:50]," "," y los primeros 58 pdf", pdf[8:50])
```

```python
#Graficar histograma

plt.figure(figsize=(6, 4))

plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title("Distribución Normal")

# Histograma de Los datos

plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)

plt.show()
```

```python
#Visualizar La PDF

plt.figure(figsize=(6, 4))

plt.plot(x, pdf, label='Normal:(media:{};ds:{})'.format(np.round(media, 2), np.round(desv_std,2)))

plt.title("Función de Densidad de Probabilidad (PDF) -;Distribución Normal")

plt.xlabel("Pesos en Kg. de los estudiantes")
plt.ylabel("Densidad de Probabilidad")

plt.legend()

plt.grid(True)

plt.show()
```

```python
#Graficar histograma

plt.figure(figsize=(6, 4))

plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')

plt.plot(x, pdf, 'k', linewidth=2)
plt.title('Distribución Normal ')

plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)

plt.show()
```

## Estudiantes que pesan menos de 72kg

```python
plt.figure(figsize=(6, 4))

#Parametros de la distribución normal

media=70 #Media
desv_std=4 #Desviación estándar
total_estudiantes=500

peso_limite=72

#Primero necesitamos calcular la probabilidad de menos que un estudiante pese
#de 72 kg utilizando la función de distribución acumulativa (CDF) de la distribución normat
prob_menos_72kg=norm.cdf(peso_limite, media, desv_std)
```

```python
#Cálculo del número de estudiantes que pesan menos de 72 kg

estudiantes_menos_72kg=total_estudiantes * prob_menos_72kg

print("Probabilidad de que un estudiante pese menos de 72 kg:", prob_menos_72kg)
print("Número de estudiantes que pesan menos de 72 kg:", estudiantes_menos_72kg)

#calcular y graficar la distribución normal

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std,1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between (x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 72 kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

## Estudiantes que pesan mas de 68kg

```python
plt.figure(figsize=(6, 4))

#Parametros de la distribución normal

media=70 #Media
desv_std=4 #Desviación estándar
total_estudiantes=500
peso_limite=68

#Primero necesitamos calcular la probabilidad de menos que un estudiante pese menos o igual a 68kg
#de 68 kg utilizando la función de distribución acumulativa (CDF) de la distribución normat
prob_menos_68kg=norm.cdf(peso_limite, media, desv_std)

# probavilidad de que un estudiante pese mas de 68kg
prob_mas_68kg = 1 - prob_menos_68kg

#Cálculo del número de estudiantes que pesan menos de 62 kg
estudiantes_mas_68kg=total_estudiantes * prob_mas_68kg

print("Probabilidad de que un estudiante pese mas de 68 kg:", prob_mas_68kg)
print("Número de estudiantes que pesan mas de 68 kg:", estudiantes_mas_68kg)
```

```python
#calcular y graficar la distribución normal

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std,1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between (x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 68 kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
