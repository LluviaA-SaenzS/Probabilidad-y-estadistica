## Ejemplo 3 histogramas de 2 grupos

1. Se generan 100 numeros aleatorios
que representan la produccion diaria de 100mts telares
2. generar otra lista con en punto anterior para una segunda empresa.
3. Con las listas se pretende determinar la media,moda,mediana,desviacion estandar
4. Graficar un histograma para cada fabrica

```python
import random
```

```python
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
  num_aleatorio = round(random.uniform(50,100),2)
  fabrica1.append(num_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  num_aleatorio = round(random.uniform(50,100),2)
  fabrica2.append(num_aleatorio)
print(fabrica2)
```

```python
# libreria para graficar
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion diaria de FABRICA1")
```

```python
plt.hist(fabrica2)
plt.title("Produccion diaria de FABRICA2")
```

```python
media = mean(fabrica1)
mediana = median(fabrica1)
moda = mode(fabrica1)
print("valores de la FABRICA 1")
print("la media es: ", media)
print("la mediana es: ",mediana)
print("la moda es: ",moda)
```

```python
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)
print("valores de la FABRICA 2")
print("la media es: ", media)
print("la mediana es: ",mediana)
print("la moda es: ",moda)
```

```python
# generar dos graficos
fig, (ax1,ax2)=plt.subplots(nrows =1, ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python
plt.hist(fabrica1, bins=10, alpha=0.4, color='g', edgecolor= "black")
plt.axvline(media, color='b', linestyle='dashed', linewidth=2, label='media')
plt.axvline(moda, color='r', linestyle='dashed', linewidth=2, label='moda')
plt.axvline(mediana, color='magenta', linestyle='dashed', linewidth=2, label='mediana')
plt.xlabel=('valor')
plt.ylabel('frecuencia')
plt.title("Produccion FABRICA1")
plt.legend()
plt.show()
```

```python
plt.hist(fabrica2, bins=10, alpha=0.4, color='g', edgecolor= "black")
plt.axvline(media2, color='b', linestyle='dashed', linewidth=2, label='media')
plt.axvline(moda2, color='r', linestyle='dashed', linewidth=2, label='moda')
plt.axvline(mediana2, color='magenta', linestyle='dashed', linewidth=2, label='mediana')
plt.xlabel=('valor')
plt.ylabel('frecuencia')
plt.title("Produccion FABRICA2")
plt.legend()
plt.show()
```
