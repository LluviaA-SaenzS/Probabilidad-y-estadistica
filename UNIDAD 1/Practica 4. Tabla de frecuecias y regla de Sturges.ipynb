## Ejemplo 3 histogramas de 2 grupos

1. Se generan 100 numeros aleatorios
que representan la produccion diaria de 100mts telares
2. generar otra lista con en punto anterior para una segunda empresa.
3. Con las listas se pretende determinar la media,moda,mediana,desviacion estandar
4. Graficar un histograma para cada fabrica

```python
import random
```

```python
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
  num_aleatorio = round(random.uniform(50,100),2)
  fabrica1.append(num_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  num_aleatorio = round(random.uniform(50,100),2)
  fabrica2.append(num_aleatorio)
print(fabrica2)
```

```python
# libreria para graficar
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion diaria de FABRICA1")
```

```python
plt.hist(fabrica2)
plt.title("Produccion diaria de FABRICA2")
```

```python
media = mean(fabrica1)
mediana = median(fabrica1)
moda = mode(fabrica1)
print("valores de la FABRICA 1")
print("la media es: ", media)
print("la mediana es: ",mediana)
print("la moda es: ",moda)
```

```python
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)
print("valores de la FABRICA 2")
print("la media es: ", media)
print("la mediana es: ",mediana)
print("la moda es: ",moda)
```

```python
# generar dos graficos
fig, (ax1,ax2)=plt.subplots(nrows =1, ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

## Dispersion de Datos

```python
desv_stand_fabrica1 = stdev(fabrica1)
print("la desviacion estandar de la FABRICA1 es ", desv_stand_fabrica1)
```

```python
desv_stand_fabrica2 = stdev(fabrica2)
print("la desviacion estandar de la FABRICA1 es ", desv_stand_fabrica2)
```

```python
plt.figure(figsize = (10,8))
plt.hist(fabrica1, bins=10, alpha=0.4, color='g', edgecolor= "black")
plt.axvline(media, color='b', linestyle='dashed', linewidth=2, label=f'media={media:.2f}')
plt.axvline(moda, color='r', linestyle='dashed', linewidth=2, label=f'moda ={moda:.2f}')
plt.axvline(mediana, color='magenta', linestyle='dashed', linewidth=2, label=f'mediana ={mediana:.2f}')

plt.axvline(media - desv_stand_fabrica1 , color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estandar={desv_stand_fabrica1:.2f}')
plt.axvline(media + desv_stand_fabrica1 , color='black', linestyle='dashed', linewidth=2)

plt.xlabel=('valor')
plt.ylabel('frecuencia')
plt.title("Produccion FABRICA1")
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.hist(fabrica2, bins=10, alpha=0.4, color='g', edgecolor= "black")
plt.axvline(media2, color='b', linestyle='dashed', linewidth=2, label=f'media={media2:.2f}')
plt.axvline(moda2, color='r', linestyle='dashed', linewidth=2, label=f'moda ={moda2:.2f}')
plt.axvline(mediana2, color='magenta', linestyle='dashed', linewidth=2, label=f'mediana ={mediana2:.2f}')

plt.axvline(media2 - desv_stand_fabrica2 , color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estandar={desv_stand_fabrica2:.2f}')
plt.axvline(media2 + desv_stand_fabrica2 , color='black', linestyle='dashed', linewidth=2)

plt.xlabel=('valor')
plt.ylabel('frecuencia')
plt.title("Produccion FABRICA2")
plt.legend()
plt.show()
```

## Tablas de Frecuencias

```python

edades =[]
for i in range (100):
  num_aleatorio = random.randint(18,70)
  edades.append(num_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
df_edades = pd.DataFrame(edades, columns=["EDAD"])
df_edades
```

```python
tabla_frec = df_edades["EDAD"].value_counts().reset_index()
tabla_frec.columns = ["EDAD","FRECUENCIA"]
tabla_frec = tabla_frec.sort_values(by = "EDAD")
tabla_frec
```

```python
tabla_frec["Frec.RELATIVA"] = tabla_frec["FRECUENCIA"]/100
tabla_frec
```

```python
tabla_frec["frec. PORCENTUAL"] = tabla_frec["FRECUENCIA"]/100*100
tabla_frec
```

```python
tabla_frec["frec. ACUMULADA"] = tabla_frec["FRECUENCIA"].cumsum()
tabla_frec
```

## Regla De Sturgles
1. Calcula numero de clases o intervalos idoneo, en los que se debe de dividir un conjunto de datos

$
      c = 1+log_2(N)
$

```python
import math
```

```python
print(edades)
```

```python
df_edades_sturg = pd.DataFrame({ "EDADES":edades})
df_edades_sturg
```

```python
n = len(edades)
k = int(1+ math.log2(n))
print("N vale= ",n,"\nK vale= ",k)
```

```python
tabla_frec_sturges = pd.cut(df_edades_sturg["EDADES"],bins=k).value_counts().reset_index()
tabla_frec_sturges
```

```python
tabla_frec_sturges.columns = ["CLASE","FRECUENCIA"]
tabla_frec_sturges
```

```python
tabla_frec_sturges = tabla_frec_sturges.sort_values(by = "CLASE")
tabla_frec_sturges
```

```python
tabla_frec_sturges["frec. RELATIVA"] = tabla_frec_sturges["FRECUENCIA"] /n
tabla_frec_sturges["frec. PORCENTUAL"] = tabla_frec_sturges["FRECUENCIA"] /n*100
tabla_frec_sturges["frec. ACUMULADA"] = tabla_frec_sturges["FRECUENCIA"].cumsum()
tabla_frec_sturges
```

```python
plt.hist(df_edades_sturg["EDADES"],bins=k, alpha=0.5, rwidth=0.85)
plt.title("histograma edades REGLA STURGLES")
plt.ylabel("FRECUENCIA")
plt.xlabel=("EDADES")
```
